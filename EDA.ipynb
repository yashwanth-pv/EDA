```
# Computes the correlation matrix to measure the linear relationship between numerical columns
corr = df.corr(numeric_only=True)
```

```
# Visualizes the correlation matrix using a heatmap to easily identify relationships between variables
sns.heatmap(corr, annot=True)
```

```
# Calculates the Interquartile Range (IQR = Q3 - Q1) to measure data spread and detect outliers
Q1 = df['death_rate'].quantile(0.25)
Q3 = df['death_rate'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['death_rate'] < Q1 - 1.5 * IQR) | (df['death_rate'] > Q3 + 1.5 * IQR)]
outliers
```

```
# Creates a contingency table to analyze the frequency distribution between two categorical variables
pd.crosstab(df['birth_rate'], df['death_rate'])
```
